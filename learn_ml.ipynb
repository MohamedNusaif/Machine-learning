```python
# @title Default title text
mi = df['Income'].mean()
mod = df['Employment_Type'].mode()[0]
```

```python
df['Income'] = df['Income'].fillna(mi)
df['Employment_Type'] = df['Employment_Type'].fillna(mod)
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Employment_Type'] = le.fit_transform(df['Employment_Type'])
df = df.loc[:, ~df.columns.duplicated()]
```

```python
df
```

```python
le.classes_
```

```python
x = df.drop(columns=['Loan_Approved'])
y = df['Loan_Approved']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
ir = LogisticRegression(max_iter=1000)
```

```python
ir.fit(x_train,y_train)
```

```python
y_pred = ir.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import classification_report
calssification_report = classification_report(y_test,y_pred)
print(calssification_report)
```
